```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('1.jpg',1)
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
image = cv2.GaussianBlur(gray_img,(5,5),10)
thresh = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)
```

```python
plt.imshow(img,cmap='gray')
```

```python
edges = cv2.Canny(thresh,50,150)
lines = cv2.HoughLinesP(thresh,1,np.pi/180,100,10)
for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(img, (x1,y1),(x2,y2),(0,0,0),2)
plt.imshow(img)
```

```python
cv2.imshow('',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
c_img = cv2.drawContours(img, contours, -1, (0,255,0), 1)
```

```python
cv2.imshow('',c_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
area = -1
for i in range(len(contours)):
    if cv2.contourArea(contours[i])>area:
        area = cv2.contourArea(contours[i])
        print('area:',area)
        print('index:',i)
max_area = area
area = -1
for i in range(len(contours)):
    if cv2.contourArea(contours[i])> area and cv2.contourArea(contours[i])<max_area:
        area = cv2.contourArea(contours[i])
        print('index:',i)
        print('area:',area)
```

```python
c_img = cv2.drawContours(img, [contours[190]], 0, (0,0,255), 2)
```

```python
cv2.imshow('',c_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
pts1 = np.float32([[371,53],[390,393],[24,388], [54,63]])
pts2 = np.float32([[300,0],[300,300],[0,300],[0,0]])
pts1
```

```python
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(300,300))
plt.subplot(121),plt.imshow(img,cmap='gray'),plt.title('Input')
plt.subplot(122),plt.imshow(dst,cmap='gray'),plt.title('Output')
plt.show()
```

```python
gray = cv2.cvtColor(dst,cv2.COLOR_BGR2GRAY)
output = cv2.bitwise_not(gray)
cv2.imshow('',output)
cv2.waitKey()
cv2.destroyAllWindows()
```
